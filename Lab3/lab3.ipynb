### **Метод Монте-Карло**

Метод Монте-Карло - чисельний метод, заснований на відтворенні великого числа реалізацій випадкового процесу, спеціально побудованого за умовами завдання. В даний час цей метод застосовується при дослідженні функціонування складних систем, до яких відносяться різноманітні виробничі та інформаційні системи, автоматизовані системи управління, багатопроцесорні обчислювальні системи, деякі біологічні, економічні системи.

При вирішенні подібних завдань раніше, без застосування EOM, джерелом випадкових чисел служили різні експерименти: кидання монети або кубика, вертіння рулетки тощо.

З ім'ям міста в князівстві Монако, відомого своїми гральними будинками, пов'язане походженням назви методу.

### **Алгоритм розв'язання задачі**

1) Генеруємо випадкове число $x_1$ в діапазоні від 0 до $\pi$, а також випадкове число $y_1$ в діапазоні від 0 до 1. Отримуємо випадкову точку всередині прямокутника з еоординатами $(x_1, y_1)$. Ця точка може потрапити в досліджуванну фігуру, а може і не потрапити.

2) Перевірка приналежності точки $(x_1, y_1)$ до досліджуваної фігурі. Якщо попадання немає, тобто $y_1>sin(x_1)$, то переходимо до пункту №1 і знову генеруємо випадкову точку з координатами $(x_2, y_2)$.

Якщо попадання є, тобто $y_1<sin(x_1)$, то необхідно зафіксувати, запам'ятати це потрапляння і знову перейти до пунтку №1. У підсумку ми повинні вирахувати число потраплянь.

Примітка: потрапляння точки точно на край фігури $y_1=sin(x_1)$ можна віднести як до одного, так і до іншого результату - це воля експериментатора і автора програми.

3) Пункт №1 слід повторити досить велике число разів, від цього, в кінцевому підсумку, залежить точність обчислень. Для даної задачі рекомендовано число повторів N = 1000 - 3000. Також **рекомендовано** провести експеримент з різними кількостями повторів та **порівняти** отриманий результат.

4) Після проведння N повторів маємо пропорцію: загальне число дослідів відповідає всій площі прямоктника, що дорівнює $\pi$, а число влучень $P$ буде відповідати невідомої площі $S$ досліджуваної фігури. Звідси $S = \frac{P\cdot\pi}{N}$

```python
from IPython.display import display, Latex
import random

def monte_carlo(amount: int, conditions: str) -> float:
    P = 0
    for i in range(amount):
        x, y = 4 * random.random(), 3 * random.random()
        try:
            if eval(conditions):
                P += 1
        except:
            print("Some problem with conditions...")
    return P/amount


conditions =    "1<=y<=3 and 0<=x<=2 or \
                    0<=y<=1 and 0<=x<=1 or \
                        1<=y<=2 and 2<=x<=4 or \
                            0<=y<=1 and 3<=x<=4"
S_all = 12
print(f"amount = 1000:\n{S_all * monte_carlo(1000, conditions)}\n\n")
print(f"amount = 2000:\n{S_all * monte_carlo(2000, conditions)}\n\n")
print(f"amount = 3000:\n{S_all * monte_carlo(3000, conditions)}")
display(Latex(r"Площа шуканої фігури : 8 $од.^2$"))
```
